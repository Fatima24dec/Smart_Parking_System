```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results = model.train(
    data="parking_dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    name="parking_yolov8"
)
```

```python
results = model("pc1.jpg")
results[0].show()

for box in results[0].boxes:
    cls_name = model.names[int(box.cls)]
    conf = float(box.conf)
    print(cls_name, round(conf, 2))
```
